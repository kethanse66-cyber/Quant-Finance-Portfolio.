```
# --- MY FIRST QUANT TOOL: RISK CALCULATOR ---

account_balance = 5000      # Total account value
risk_per_trade = 0.01       # 1% risk per trade
stop_loss_pips = 20         # Stop-loss distance

# Calculate risk and position size
dollar_risk = account_balance * risk_per_trade
position_size = dollar_risk / stop_loss_pips

print("Total Account:", account_balance)
print("Risk per Trade:", dollar_risk)
print("Suggested Position Size:", position_size)
```
