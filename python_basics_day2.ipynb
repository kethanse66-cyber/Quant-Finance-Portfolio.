```
# Python Basics - Day 2
# Topics: Variables, Lists, if-else, for loop

# Variables
price = 100
ema = 95

# if-else decision
if price > ema:
    trend = "UP"
else:
    trend = "DOWN"

print("Trend:", trend)

# List of prices
prices = [100, 102, 101, 105]

# Loop through prices
for p in prices:
    if p > ema:
        print(p, "above EMA")
    else:
        print(p, "below EMA")
```
